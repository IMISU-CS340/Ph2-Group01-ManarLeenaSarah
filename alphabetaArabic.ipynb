#### 3.1. Alpha beta pruning

هو نوع من البحث العدائي والذي هو تطوير خوارزمية Minimax. خوارزمية Minimax هي أيضًا نوع من البحث العدائي الذي تعمل على النحو التالي:

يعمل اللاعب على زيادة أدائه ، لكن الخصم سيعمل على تقليل أداء اللاعب إلى الحد الأدنى ، وتقوم خوارزمية Minimax بإنشاء شجرة البحث بأكملها. تحتوي بعض المشكلات على مساحة بحث ضخمة ، تأتي Alphabeta مع فكرة التقليم لتقليل مساحة البحث وضمان نفس الناتج من Minimax.
اذا ما المقصود بالتقليم؟ إنه إجراء قطع العقدة غير الضرورية التي لن تؤثر على النتيجة النهائية من شجرة البحث ، يتم استخدامه لتقليل مساحة البحث. سنرى التقليم في الحالة الأولية المختارة.
تُعرف الخوارزمية باسم Alphabeta لأنه في التقليم يعتمد على متغيرين ؛ ألفا وبيتا:

ألفا: أفضل خيار (أعلى قيمة) وجدناه حتى الآن في أي مرحلة على طول مسار Maximizer. القيمة الأولية لـ alpha هي -∞.

بيتا: أفضل خيار (أقل قيمة) وجدناه حتى الآن في أي مرحلة على طول مسار Minimizer. القيمة الأولية لبيتا هي + ∞. [3]

يتم التقليم إذا تحقق الشرط التالي: alpha> = beta.


<a id="eva"></a>
#### 3.2. Evaluation function

أو تسمى أيضا بـUtallity function.
وهي دالة تحدد القيم للأفرع النهائية على الوجه التالي:


إذا فاز X تقوم بإرجاع رقم> 0 .


إذا لم يفز أحد (تعادل) تقوم بإرجاع 0.


إذا ربح O تقوم بإرجاع رقم <0.


بعض المشاكل معقدة ولديها مساحة بحث ضخمة, لا نصل إلى الحالات النهائية ولكننا ننشئ شجرة البحث إلى عمق معين (يسمى horizon) ونستخدم utality function مع heuristic  لتقييم كل حالة مختلفة في مستوى horizon ، إحدى طرق حساب القيم في مستوى  هي f (n) = heuristic of max - heuristic of min
heuristic تمثل مجموع عدد الأعمدة والصفوف المتاحة للاعب لكي يفوز في حالة معينة. في الصورة التالية مثال على طريقة حساب heuristic.

لأن مشكلتنا تحتوي على مساحة بحث صغيرة يمكننا الوصول إلى الحالات النهائية ولا حاجة لاستخدام قإلى تقييد البحث حتى عمق معين.

#### 3.3. Pseudocode

function ALPHA-BETA(state) returns an action

   inputs: state, current state in game
   
   v=MAX-VALUE(state, - ∞ , +∞)
   
   return the action in SUCCESSORS(state) with value v
   
function MAX-VALUE(state,alpha , beta) returns a utility value

   if TERMINAL-TEST(state) then return UTILITY(state)
   
   v = - ∞
   
   for a,s in SUCCESSORS(state) do
   
      v = MAX(v,MIN-VALUE(s, alpha , beta))
      
     if alpha ≥ beta then return v
     
     alpha = MAX(alpha ,v)
     
   return v
   
function MIN-VALUE(state, alpha , beta) returns a utility value

   if TERMINAL-TEST(state) then return UTILITY(state)
   
   v = + ∞
   
   for a,s in SUCCESSORS(state) do
   
      v = MIN(v,MAX-VALUE(s, alpha , beta))
      
      if beta ≤ alpha then return v
     
      beta = MIN(beta ,v)
      
   return v

#### 3.4 implementation


```python
#code
#implement the initial state as matrix
state=[['o','o','x'],
       ['e','x','e'],
       ['o','x','e']]

#saved is an array that will hold the coordinates for the best move
saved=[]

def alphabeta(state,player,alpha,beta):
    
    #terminal states
    
    for r in range(len(state)):
        for c in range(len(state[r])):
            #three connected in one column
            if r+2<len(state) and state[r][c]==state[r+1][c]==state[r+2][c]:
                if state[r][c]=='x':
                    return 10
                elif state[r][c]=='o':
                    return -10
                #three connected in one row
            if c+2<len(state[r]) and state[r][c]==state[r][c+1]==state[r][c+2]:
                if state[r][c]=='x':
                    return 10
                elif state[r][c]=='o':
                    return -10
                #three conected in the main diagonal
            if r+2<len(state) and c+2<len(state[r]) and state[r][c]==state[r+1][c+1]==state[r+2][c+2]:
                if state[r][c]=='x':
                    return 10
                elif state[r][c]=='o':
                    return -10
                #three connected in secondary diagonal
            if c-2>-1 and r+2<len(state) and state[r][c]==state[r+1][c-1]==state[r+2][c-2]:
                 if state[r][c]=='x':
                    return 10
                 elif state[r][c]=='o':
                    return -10
                
    #game ending test   
    
    complete=True
    for r in range(len(state)):
         for c in range(len(state[r])):
                if state[r][c]=='e':
                    complete=False
         
    if complete:
        return 0  
    
    #max player
    if player:
        for r in range(len(state)):
            for c in range(len(state[r])):
                if state[r][c]=='e':
                    newstate=state
                    newstate[r][c]='x'
                    value=alphabeta(newstate,False,alpha,beta)
                    if alpha<value:
                        saved.append((r,c))
                        alpha=value
                    if beta <= alpha:
                        break
        return alpha
    else:#min player
        for r in range(len(state)):
            for c in range(len(state[r])):
                if state[r][c]=='e':
                    newstate=state
                    newstate[r][c]='o'
                    value=alphabeta(newstate,True,alpha,beta)
                    if beta>value:
                        saved.append((r,c))
                        beta=value
                    if beta <= alpha:
                        break
        return beta
    
print("النقاط هي :",alphabeta(state,True,-1000000,1000000))
if len(saved)>0:
    print('أفضل خطوة ستكون في الإحداثية: ',saved[len(saved)-1])
print('الحالة النهائية:')
print(state)

```
